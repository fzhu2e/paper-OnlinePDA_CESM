# Create restoring target for the standard restoring (SR) run

```python
import os
import x4c
import xarray as xr
os.chdir('/glade/work/fengzhu/Projects/paleoWeather_OnlineDA_method/notebooks')
print(x4c.__version__)
```

## Field

```python
dirpath = '/glade/campaign/univ/uazn0034/fengzhu/CESM_output/timeseries/b.e13.B1850C5CN.ne30_g16.icesm131_ihesp.pW.Plio_350ppm'
case = x4c.Timeseries(dirpath, grid_dict={'atm': 'ne30np4', 'ocn': 'g16'}, cesm_ver=1)
timespan = (721, 780)
case.load('TEMP', timespan=timespan)
case.ds['TEMP']
```

```python
da_prior = case.ds['TEMP']['TEMP'].isel(time=slice(-50*12, None))
da_prior_anom = da_prior - da_prior.mean('time')
for R in [0.01, 0.25, 1.00]:
    ds_target = xr.open_dataset(f'../recons/recon_Plio400ppm_R{R:.2f}_loc10000.nc')
    da_target = da_prior_anom + ds_target.isel(time=0)['TEMP'].mean('ens')
    da_target_climo = da_target.x.climo
    ds_out = xr.Dataset()
    for i in range(1, 13):
        da_tmp = da_target_climo.sel(time=i).rename({'z_t': 'depth', 'nlon': 'X', 'nlat': 'Y'}).drop_vars(['ULAT', 'ULONG'])
        da_tmp.attrs.clear()
        ds_out[f'TEMPERATURE{i:02d}'] = da_tmp
    ds_out.to_netcdf(f'./restoring_target_ppeDA_R{R}_g16_TEMP_climo_350ppm.nc')
```

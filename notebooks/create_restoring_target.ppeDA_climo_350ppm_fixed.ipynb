```python
ds_target = xr.open_dataset('./restoring_target_ppeDA_R0.01_g16_TEMP_climo_350ppm.nc')
ds_target
```

```python
dirpath = '/glade/campaign/univ/uazn0034/fengzhu/CESM_output/timeseries/b.e13.B1850C5CN.ne30_g16.icesm131_ihesp.pW.Plio_350ppm.restore_ppeDA_R0.01_tau0.05_climo'
case = x4c.Timeseries(dirpath, grid_dict={'atm': 'ne16np4'}, cesm_ver=1)
TEMP_climo_init = case.calc('TEMP:climo')
```

```python
ds_out = xr.Dataset()
for i in range(1, 13):
    da_target = ds_target[f'TEMPERATURE{i:02d}']-TEMP_climo_init.sel(time=i).rename({'z_t': 'depth', 'nlon': 'X', 'nlat': 'Y'}).drop_vars(['ULAT', 'ULONG'])
    da_target.attrs.clear()
    ds_out[f'TEMPERATURE_diff{i:02d}'] = da_target

ds_out.to_netcdf('./restoring_target_ppeDA_R0.01_g16_TEMP_climo_350ppm_fixed.nc')
ds_out
```
